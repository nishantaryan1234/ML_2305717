```python
import math
points = [
    (1,1),(1,2),(2,1),(2,2),(3,2),   # cluster 1
    (8,8),(8,9),(9,8),(9,9),(10,8),   # cluster 2
    (5,5)                             # noise
]
eps = 1.5
minPts = 4
labels = [0]*len(points)      # 0 = unvisited
cluster_id = 0
def distance(p1,p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)
def region_query(i):
    neighbors = []
    for j in range(len(points)):
        if distance(points[i],points[j]) <= eps:
            neighbors.append(j)
    return neighbors
def expand_cluster(i,neighbors,cid):
    labels[i] = cid
    k = 0
    while k < len(neighbors):
        n = neighbors[k]
        if labels[n] == -1:
            labels[n] = cid
        elif labels[n] == 0:
            labels[n] = cid
            new_neighbors = region_query(n)
            if len(new_neighbors) >= minPts:
                neighbors += new_neighbors
        k += 1
for i in range(len(points)):
    if labels[i] != 0:
        continue
    neighbors = region_query(i)
    if len(neighbors) < minPts:
        labels[i] = -1
    else:
        cluster_id += 1
        expand_cluster(i,neighbors,cluster_id)
point_type = []
for i in range(len(points)):
    if labels[i] == -1:
        point_type.append("Noise")
    else:
        neighbors = region_query(i)
        if len(neighbors) >= minPts:
            point_type.append("Core")
        else:
            point_type.append("Boundary")
print("Point\tCluster\tType")
for i in range(len(points)):
    print(points[i], "\t", labels[i], "\t", point_type[i])
```

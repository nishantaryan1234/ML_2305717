## Divisive Hierarchial chustering

```python
import math

def euclidean_distance(p1, p2):
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(p1, p2)))

def find_farthest_points(data):
    max_dist = -1
    p1, p2 = None, None

    for i in range(len(data)):
        for j in range(i + 1, len(data)):
            dist = euclidean_distance(data[i], data[j])
            if dist > max_dist:
                max_dist = dist
                p1, p2 = data[i], data[j]

    return p1, p2

def split_cluster(cluster):
    if len(cluster) <= 1:
        return [cluster]

    p1, p2 = find_farthest_points(cluster)

    cluster1 = []
    cluster2 = []

    for point in cluster:
        if euclidean_distance(point, p1) < euclidean_distance(point, p2):
            cluster1.append(point)
        else:
            cluster2.append(point)

    return [cluster1, cluster2]

def divisive_clustering(data, k):
    clusters = [data]

    while len(clusters) < k:
        largest_cluster = max(clusters, key=len)
        clusters.remove(largest_cluster)

        new_clusters = split_cluster(largest_cluster)

        clusters.extend(new_clusters)

    return clusters

data = [
    [1,2], [2,1], [2,2],
    [8,8], [9,8], [8,9],
    [25,30], [24,29]
]

k = 3

clusters = divisive_clustering(data, k)
print("Initial Data Set")
print(data)
print("Clusters:")
for i, cluster in enumerate(clusters):
    print("Cluster", i+1, ":", cluster)
```
